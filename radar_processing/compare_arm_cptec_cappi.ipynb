```python
from datetime import datetime
import numpy as np
import netCDF4
from matplotlib import colors
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
from matplotlib.pyplot import tight_layout


def plot_panel(arm, cptec, date):
    # Create figure and axes
    fig, axs = plt.subplots(1,2)
    fig.set_size_inches(7,4.5)

    # Display the image
    ax = axs[0].imshow(arm.variables["DBZc"][0,5,:,:], vmin=-10, vmax=80, cmap=plt.get_cmap("gist_ncar"))
    # Create a Rectangle patch
    rect = patches.Rectangle((45, 45), 150, 150, linewidth=1, edgecolor='r', facecolor='none')
    # Add the patch to the Axes
    axs[0].add_patch(rect)
    # Title
    axs[0].set(title="ARM CAPPI - 3 km")

    # Display the image
    axs[1].imshow(cptec[1,:,:], vmin=-10, vmax=80, cmap=plt.get_cmap("gist_ncar"))
    # Create a Rectangle patch
    rect = patches.Rectangle((100, 100), 300, 300, linewidth=1, edgecolor='r', facecolor='none')
    # Add the patch to the Axes
    axs[1].add_patch(rect)
    # Title
    axs[1].set(title="CPTEC CAPPI - 3 km")

    # Colorbar
    cbar = fig.colorbar(ax, ax=axs, orientation='horizontal', fraction=.1)
    cbar.set_label("dBZ")

    # Main title
    fig.suptitle(date)

    # plt.show()

    plt.savefig("tests_tathu/figs/arm_cptec_panel_" + date.strftime("%Y%m%d_%H%M%S") + ".png", dpi=300)

    plt.clf()
    plt.cla()

    print("Plotted! " + str(date))
```

```python
# arm_file = "../data/radar/sipam_manaus/arm_cappi/2015-04/20150403/sbmn_cappi_20150403_084807.nc"
# cptec_file = "../data/radar/sipam_manaus/cptec_cappi/2015/cappi_201504030848.dat"
for arm_file, cptec_file in zip (arm_files, cptec_files):
    # Read ARM data
    arm = netCDF4.Dataset(arm_file)
    # Read CPTEC data
    cptec = np.fromfile(cptec_file, dtype="float32").reshape(15,500,500)
    cptec[cptec == -99.] = np.nan
    # Get date
    date = datetime.strptime(arm_file[-18:-3], "%Y%m%d_%H%M%S")
    # Plot
    plot_panel(arm, cptec, date)
```
