# FunÃ§Ã£o para juntar CSs com trajetÃ³rias

```python
def join_hst_sys(hst, sysfilt):
    geo_sysfilt = gpd.GeoDataFrame(sysfilt.to_frame().T, geometry=sysfilt.to_frame().T.geom, crs="EPSG:4326").to_crs(crs=3857)
    hst = hst.loc[hst.date_hours >= -6]
    inter = abs(geo_sysfilt.iloc[0].timestamp - hst.date)
    hst = hst.loc[inter[inter <= pd.Timedelta(6, "H")].index].to_crs(crs=3857)
    pts_within = hst.sjoin_nearest(geo_sysfilt, max_distance=10000, distance_col="distance").reset_index().to_crs(crs="EPSG:4326")
    # pts_within = hst.sjoin_nearest(geo_sysfilt, max_distance=0.00001, distance_col="distance").reset_index().to_crs(crs="EPSG:4326")
    pts_within["inter"] = abs(pts_within.timestamp - pts_within.date)
    pts_within2 = pts_within.loc[pts_within.groupby("date_hours")["inter"].idxmin()].loc[pts_within.inter < pd.Timedelta(6, "H")]
    # print(pts_within2)
    if not pts_within2.empty:
        result = pts_within2.loc[pts_within2["date_hours"].idxmax()]
    else:
        result = pts_within2.append(pd.Series(0, index=pts_within2.columns), ignore_index=True).iloc[0]
    print("done!")
    return result
```
